```python
import fastf1.plotting
import fastf1
import matplotlib.pyplot as plt


fastf1.plotting.setup_mpl(misc_mpl_mods=False)
session = fastf1.get_session(2023, 1, 'R')
session.load(telemetry=False, weather=False)

fig, ax = plt.subplots(figsize=(8.0, 4.9))

a = []
b = []
c = []
for drv in session.drivers:
    drv_laps = session.laps.pick_driver(drv)

    abb = drv_laps['Driver'].iloc[0]
    color = fastf1.plotting.driver_color(abb)

    ax.plot(drv_laps['LapNumber'], drv_laps['Position'],
            label=abb, color=color)
    pos_gained = drv_laps['Position'].iloc[0] - drv_laps['Position'].iloc[-1]
    a.append(abb)
    b.append(pos_gained)
    c. append(drv_laps['Position'].iloc[0])

ax.set_ylim([20.5, 0.5])
ax.set_yticks([1, 5, 10, 15, 20])
ax.set_xlabel('Lap')
ax.set_ylabel('Position')

ax.legend(bbox_to_anchor=(1.0, 1.02))
plt.tight_layout()

plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Constants for the simulation
NUM_SIMULATIONS = 1000
NUM_LAPS = 78  # Example for Monaco GP
TYRE_COMPOUNDS = ['soft', 'medium', 'hard']
WEATHER_CONDITIONS = ['dry', 'wet']
OVERTAKING_PROBABILITY_BASE = 0.4
NUM_CARS = 20
Base_LAP = 97
PENALTY_PER_POSITION = 0.05
#Exponential Decay Penalty: Instead of a linear penalty per position, use an exponential decay model. This would simulate the reality that the difference between each position decreases as you move down the grid. For example, the difference between 1st and 2nd is more significant than between 19th and 20th.
decay_factor = 0.005
def tyre_degradation_curve(lap, compound):
    if compound == 'soft':
        return  (model3[4]* lap**4) + (model3[3]* lap**3) + (model3[2]* lap**2) + (model3[1]* lap) + model3[0]
    elif compound == 'medium':
        return  (model2[4]* lap**4) + (model2[3]* lap**3) + (model2[2]* lap**2) + (model2[1]* lap) + model2[0]
    elif compound == 'hard':
        return  (model1[4]* lap**4) + (model1[3]* lap**3) + (model1[2]* lap**2) + (model1[1]* lap) + model1[0]
    else:
        raise ValueError('Invalid tyre compound')

def weather_effect(weather):
    return 1.0 if weather == 'wet' else 1.0

def time_to_pass(driver_position, lap):
    # Simplified model for time to pass the car ahead
    return 0.5 * (1 / driver_position) * lap

def pit_stop_time(lap, has_pitted):
    # Assuming a fixed time for pit stops
    return 27 if lap in has_pitted else 0

# discounted fuel effect as fuel is same for all teams and is note refilled
```

```python
def simulate_race_performance(quali_position, tyre_strategy, weather_conditions):
    lap_times = []
    has_pitted = []
    current_tyre = tyre_strategy[0]
    x = 0
    positions = np.arange(1, NUM_CARS + 1) 
    for lap in range(NUM_LAPS):
        for car in range(NUM_CARS):
            if lap in tyre_strategy:
                current_tyre = tyre_strategy[lap]
                if lap not in has_pitted:
                    has_pitted.append(lap)
                else:
                    x = x+ 27
                
            base_lap_time = 95 # change to quickest time in previous race for each compound
            lap_time = tyre_degradation_curve(lap, current_tyre)
            degradation = (lap_time-base_lap_time)/base_lap_time
            lap_time += Base_LAP * (np.exp(-quali_position / decay_factor))
            #lap_time += quali_position * PENALTY_PER_POSITION
            #lap_time *= weather_effect(weather_conditions[lap])
            #lap_time += time_to_pass(quali_position, lap)
            #lap_time += pit_stop_time(lap, has_pitted)
            lap_times.append(lap_time)

            if car > 0 and np.random.rand() < (OVERTAKING_PROBABILITY_BASE - degradation):
                positions[car], positions[car - 1] = positions[car - 1], positions[car]
                #lap_time += time_to_pass(positions, lap)

    return sum(lap_times) +x , positions, lap_times 
```

```python
import numpy as np

TYRE_COMPOUNDS = ['soft', 'medium', 'hard']

# Function to generate tyre strategy
def generate_tyre_strategy():
    tyre_strategy = {}

    # Start with a random choice
    first_choice = np.random.choice(TYRE_COMPOUNDS)
    tyre_strategy[0] = first_choice

    # Ensure the second choice is different
    second_choice = np.random.choice([c for c in TYRE_COMPOUNDS if c != first_choice])
    tyre_strategy[1] = second_choice

    # For the rest, choose randomly but ensure no more than two of each
    compound_counts = {c: (2 if c in [first_choice, second_choice] else 1) for c in TYRE_COMPOUNDS}
    
    for lap in range(2, NUM_LAPS):
        available_compounds = [c for c, count in compound_counts.items() if count < 2]
        if not available_compounds:  # All compounds have been used twice
            break
        choice = np.random.choice(available_compounds)
        tyre_strategy[lap] = choice
        compound_counts[choice] += 1

    return tyre_strategy

# Example of usage
```

```python
def monte_carlo_simulation():
    results = []
    for _ in range(NUM_SIMULATIONS):
        quali_position = np.random.randint(1, 21)
        weather_conditions = np.random.choice(WEATHER_CONDITIONS, NUM_LAPS)
        #tyre_strategy = {0: np.random.choice(TYRE_COMPOUNDS, replace= False)}
        tyre_strategy = generate_tyre_strategy()
        for lap in range(1, NUM_LAPS):
            if np.random.rand() < 0.025:  # Chance of changing tyres
                tyre_strategy[lap] = np.random.choice(TYRE_COMPOUNDS,replace= False)
        total_time, positions, lap_times = simulate_race_performance(quali_position, tyre_strategy, weather_conditions)
        results.append((quali_position, tyre_strategy, total_time, positions, lap_times))
    return results

simulation_results = monte_carlo_simulation()
```

```python
def analyze_results(simulation_results):
    df = pd.DataFrame(simulation_results, columns=['Qualifying Position', 'Tyre Strategy', 'Total Race Time','positions','laptimes',])#'scenario_likelihood'])
    df['LastPosition'] = df['positions'].apply(lambda x: x[-1])  
    optimal_strategies = df.groupby('Qualifying Position')['LastPosition'].idxmin()
    best_strategies = df.loc[optimal_strategies]
    return best_strategies

best_strategies = analyze_results(simulation_results)
best_strategies
#print("Optimal Strategies by Qualifying Position:\n", best_strategies)
#first_row_tyre_strategy = best_strategies.loc[0, 'Tyre Strategy']
#first_row_tyre_strategy
```

```python
def plot_strategies(best_strategies):
    plt.figure(figsize=(10, 6))
    plt.scatter(best_strategies['Qualifying Position'], best_strategies['Total Race Time'])
    plt.xlabel('Qualifying Position')
    plt.ylabel('Total Race Time')
    plt.title('Optimal Race Strategies by Qualifying Position')
    plt.show()

plot_strategies(best_strategies)
```
